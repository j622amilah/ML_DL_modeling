```python
def intpos(x, y, n):
    front = n-y-1
    back = y
    left = x
    right = n-x-1
    
    return front, back, left, right
```

```python
#------------------
# Simple behavior:
#------------------
# bouger_x direction
# bouger_y direction

def straight(b_x, b_y, x, y, n):
    
    front, back, left, right = intpos(x, y, n)
    # print('front : ', front)
    # print('back : ', back)
    # print('left : ', left)
    # print('right : ', right)
    
    # limitations to command
    # print('before b_x : ', b_x)
    if np.sign(b_x) == -1: # gauche
        if np.abs(b_x) > left:
            b_x = -left # new limited command
    elif np.sign(b_x) == 1: # droit
        if b_x > right:
            b_x = right # new limited command
    # print('after b_x : ', b_x)
    
    # print('before b_y : ', b_y)
    if np.sign(b_y) == -1: # derriere
        if np.abs(b_y) > back:
            b_y = -back # new limited command
    elif np.sign(b_y) == 1: # avant
        if b_y > front:
            b_y = front # new limited command
    # print('after b_y : ', b_y)
    
    # no limitations
    xnew = x + b_x
    ynew = y + b_y
    # print('xnew : ', xnew)
    # print('ynew : ', ynew)
    
    return xnew, ynew
```

```python
#------------------
# Simple behavior:
#------------------
# bouger_haut_droit

def diagonal_hd(b_hd, x, y, n): # up_right_diag
    
    front, back, left, right = intpos(x, y, n)
    # print('front : ', front)
    # print('back : ', back)
    # print('left : ', left)
    # print('right : ', right)
    
    # limitations to command
    # print('before b_hd : ', b_hd)
    
    # print('np.sign(b_hd) : ', np.sign(b_hd))
    
    if np.sign(b_hd) == -1: # gauche
        # print('b_hd is negative : ')
        # print('np.abs(b_hd) : ', np.abs(b_hd))
        # print('left : ', left)
        if np.abs(b_hd) > left or np.abs(b_hd) > back:
            b_hd = -np.min([left, back]) # new limited command
            # print('-np.min([left, back]) : ', -np.min([left, back]))
    elif np.sign(b_hd) == 1: # droit
        # print('b_hd is positive : ')
        # print('b_hd : ', b_hd)
        if b_hd > right or b_hd > front:
            b_hd = np.min([right, front]) # new limited command
            # print('np.min([right, front]) : ', np.min([right, front]))
    # print('after b_hd : ', b_hd)
    
    # no limitations
    xnew = x + b_hd
    ynew = y + b_hd
    # print('xnew : ', xnew)
    # print('ynew : ', ynew)
    
    return xnew, ynew
```

```python
#------------------
# Complex behavior:
#------------------
def diagonal_bd(b_bd, x, y, n): # down_right_diag
    
    front, back, left, right = intpos(x, y, n)
    # print('front : ', front)
    # print('back : ', back)
    # print('left : ', left)
    # print('right : ', right)
    
    # print('before b_bd : ', b_bd)
    if np.sign(b_bd) == -1:   # we want to move bas_droit
        if np.abs(b_bd) > left or np.abs(b_bd) > front:
            b_bd = -np.min([left, front]) # new limited command
    elif np.sign(b_bd) == 1: # we want to move the opposite of physical bas_droit
        if b_bd > right or b_bd > back:
            b_bd = np.min([right, back]) # new limited command
    # print('after b_bd : ', b_bd)
    
    # no limitations
    xnew = x + b_bd
    ynew = y - b_bd
    # print('xnew : ', xnew)
    # print('ynew : ', ynew)
    
    return xnew, ynew
```
